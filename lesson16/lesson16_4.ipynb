```python
import pandas as pd
#將DataFrame當作資料庫來操作
df = pd.read_csv('world.csv')
```

```python
df1 = df[['洲名','國家','日期','總確診數','新增確診數','總人口數','解封指數']]
```

```python
df1['日期'] = pd.to_datetime(df1['日期'])
```

```python
#實體方法query()
#query()內放的是字串

df1_taiwan = df1.query("國家=='台灣'")
df1_taiwan
```

```python
df1_japan = df1.query("國家=='日本'")
df1_japan
```

```python
df1_japan.query("日期.dt.year == 2021 and (日期.dt.month == 5) or (日期.dt.month == 8)")
df1_japan
```

```python
df1_japan.query("日期.dt.year == 2021")
```
