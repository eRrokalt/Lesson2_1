```python
import numpy as np 
one_store = np.random.randint(100,10000,size=(30,))
one_store
```

```python
oneStorecum = one_store.cumsum()
oneStorecum
```

```python
import matplotlib.pyplot as plt
figure = plt.figure(figsize=(10,5))
axes = figure.add_subplot(111)
axes.plot(range(1,31),oneStorecum,marker = 'o',linestyle = '-',color = 'r',label = 'Cumulative Revenue')
axes.set_title('Cumulative Revenue Over time')
axes.set_xlabel('day')
axes.set_ylabel('Revenue')
axes.legend()
axes.grid()




```
